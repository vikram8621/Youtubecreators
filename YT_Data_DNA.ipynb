```python
translator = Translator(to_lang='en')
```

```python
def translate_to_english(text):
    try:
        translation = translator.translate(text)
        return translation
    except Exception as e:
        return text
```

```python
import pandas as pd
import chardet
import re

# Function to detect and set the correct encoding
def detect_and_set_encoding(file_path):
    with open(file_path, 'rb') as rawdata:
        result = chardet.detect(rawdata.read(10000))
    return result['encoding']

# Function to clean and normalize text
def clean_and_normalize_text(text):
    # Replace non-alphanumeric characters with spaces
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    # Remove extra spaces and strip leading/trailing spaces
    text = ' '.join(text.split())
    return text

# File path to your CSV file
csv_file = 'Datadna_YT_Nov23.csv'

# Detect and set the correct encoding
encoding = detect_and_set_encoding(csv_file)

# Read the CSV file with the detected encoding
df = pd.read_csv(csv_file, encoding=encoding)

# Clean and normalize the text in the 'Name' column
df['Youtuber'] = df['Youtuber'].apply(clean_and_normalize_text)

# Save the cleaned data to a new CSV file
cleaned_csv_file = 'cleaned_data.csv'
df.to_csv(cleaned_csv_file, index=False)

print(f"Data cleaned and saved to '{cleaned_csv_file}'.")
```
